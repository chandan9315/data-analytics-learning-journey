```python
# Factorial
def factorial(n):
    fact = 1
    for i in range(1, n+1):
        fact *= i
    return fact

print(factorial(5))
```

```python
# Reverse string
def reverse_string(s):
    return s[::-1]

print(reverse_string("Python"))
```

```python
def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = 0
    for i in s:
        if i in vowels:
            count += 1
    return count

print(count_vowels("Hello coders"))
```

```python
def fibonacci(k):
    a, b = 0, 1
    for i in range(k):
        print(a, end=" ")
        a, b = b, a+b

fibonacci(7)
```

```python
def view():
    for i in range(0,5):
        yield i              # yield return the value when value is generate and give.

h=list(view())
print(h)
```

```python
k=[22,33,4,45,5,6,7,7,8,5,4,3,45,56,6,7,6,45,4]
def show():
    for i in k:
        yield(i)

a=list(show())
print(a)
```

```python
# 7. Positional Argument Function
def intro(name, age):
    print("Name:", name)
    print("Age:", age)

intro("Chandan", 20)
```

```python
# 8. Anonymous Function (Lambda Function
square = lambda x: x*x
print(square(12))
```

```python
# 9. Return Type Function
def addition(a, b):
    return a + b

result = addition(15,27)
print(result)
```

```python
# 10. Recursion Function
def facto(n):
    if n == 1:
        return 1
    else:
        return n * facto(n-1)

print(facto(5))
```
