```python
# Decorator: 
# A decorator in Python is a function that takes another function as an argument, adds extra functionality,
# and returns a new function (a "decorated" version) without changing the original function's source code.
# This is a powerful feature for adding reusable behaviors like logging, authentication, or timing to multiple functions in a clean, readable way
```

```python
# A decorator in Python is a design pattern used to modify or extend the behavior of a function or method without changing its actual code.
# Technically, decorators are higher-order functions (functions that take another function as input and return a new function)
```

```python
def hello():
    print("Hello")

def my_decorator(func):
    def wrapper():
        print("Before function")
        func()
        print("After function")
    return wrapper

@my_decorator
def hello():
    print("Hello")
```
