```python
# Recursion:
# Recursion in Python is a programming technique where a function calls itself to solve a problem by breaking it down into smaller,
# simpler subproblems. This process continues until a basic, solvable condition is m
```

```python
def fact(x):
    if(x==1):
        return 1
    else:
        return x*fact(x-1)

t=fact(5)
print(t)
```

```python
def sum(x):
    if (x==1):
        return 1
    else:
        return x + sum(x-1)

t=sum(10)
print("sum of 1 to 10 : ",t)
```

```python
def sum_n(x):
    if (x==11):
        return 0
    else:
        return x + sum_n(x+1)

t=sum_n(1)
print("sum of 1 to 10 : ",t)
```
