```python
# Recursion:
# Recursion in Python is a programming technique where a function calls itself to solve a problem by breaking it down into smaller,
# simpler subproblems. This process continues until a basic, solvable condition is m
```

```python
def fact(x):
    if(x==1):
        return 1
    else:
        return x*fact(x-1)

t=fact(5)
print(t)
```
