```python
one=pd.DataFrame({
    'id':[1,2,3,4,5],
    'Name':['Akash','Anshu','Aadil','Badal','Chandan'],
    'subject_id':['sub1','sub2','sub3','sub4','sub5']})
one
```

```python
two=pd.DataFrame({
    'id':[1,2,3,14,15],
    'Name':['Anil','Aditya','Aadil','Bhupesh','Deepak'],
    'subject_id':['sub1','sub2','sub3','sub4','sub5']})
two
```

```python
# joins in pandas
# Left join  (It can show the full left side data and compair with right side data.)
# Right join  (It can show the full right side data and compair with left side data.)
# Inner join  (It can show the common data in both left and right side data.)
# Outer join  (It can show the overall compairsion of left or right both side data at a time.)
```

```python
# By default is apply inner join.
print(pd.merge(one,two,on='id'))
```

```python
print(pd.merge(one,two,on='Name',how='left'))
```

```python
print(pd.merge(one,two,on='id',how='right'))
```

```python
print(pd.merge(one,two,on='id',how='inner'))
```

```python
print(pd.merge(one,two,on='id',how='outer'))
```

```python
# apply these join methods behalf of two coloumns.
print(pd.merge(one,two,on=['id','Name']))
```

```python
print(pd.merge(one,two,on=['id','Name'],how='outer'))
```

```python
# Group by: A groupby operation involves some combination of splitting the object, applying a function, and combining the results. 
#           This can be used to group large amounts of data and compute operations on these groups. 
#            Parameters: bymapping, function, label, pd.Grouper or list of such. Used to determine the groups for the groupby.
```

```python
df
```

```python
h=df.groupby('Department')
h
```

```python
for i,j in h:
    print(i)
    print(j)
```

```python
b=h['Salary'].sum()
b
```

```python
type(b)
```

```python
# Now we can change this data type in dataframe for performing the pandas methods.
k=pd.DataFrame(b)
k
```

```python
k.sort_values('Salary',ascending=False)
```
