```python
d={'id':3456,'name':'chandan','age':21,'contact':9312436787}
print(d.keys())     # to find the all keys of dictionary.
print(d.values())    # to fing the all values of dict.
print(d.items())         # to find the all keys and values
len(d.items())          # to print the length
```

```python
d={'id':3456,'name':'chandan','age':21,'contact':9312436787}
for i in d.items():
    print(i)
for i,j in d.items():
    print(i,"-",j)
```

```python
student = {"name": "Chandan", "age": 21, "course": "BCA"}

student.pop("age")
del student["course"]

print(student)
```

```python
student = {"name": "Chandan", "age": 21, "course": "BCA"}

student.pop("age")                    # to delete the element using keys.
del student["course"]

print(student.get('name'))            # to get the element using keys.
```
