```python
# Methods for Modifying Lists
# append(): Adds a single element to the end of the list. Example: my_list.append(4) results in [1, 2, 3, 4].
# extend(): Adds the elements of another iterable (like another list) to the end of the current list.
#             Example: my_list.extend([4, 5]) results in [1, 2, 3, 4, 5].
# insert(): Adds an element at a specified position/index. Example: my_list.insert(1, 5) results in [1, 5, 2, 3].
# remove(): Removes the first occurrence of a specified value. A ValueError is raised if the item is not found. 
#             Example: my_list.remove(2) results in [1, 3, 2] from [1, 2, 3, 2].
# pop():    Removes and returns the element at a specified index (defaults to the last item if no index is provided). 
#           Example: popped_element = my_list.pop(1) removes the item at index 1 and returns it.
# clear(): Removes all elements from the list, making it empty. Example: my_list.clear() results in []. 

# Methods for Reordering and Copying
# sort():    Sorts the list in ascending order (in-place modification). You can use reverse=True as a parameter to sort in descending order.
# reverse(): Reverses the current order of the elements in the list in place.
# copy():    Returns a shallow copy of the list, so modifications to the new list do not affect the original. 
    
# Methods for Information Retrieval
# count(): Returns the number of elements with the specified value.
# index(): Returns the index of the first element with the specified value. A ValueError is raised if the value is not present
```
