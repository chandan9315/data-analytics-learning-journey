```python
## Tuple:
# A tuple in Python is an immutable ordered collection of elements.
# tuple denoted by () bracket.
# It is immutable.
# It can store the data on index.
# tuple accept duplicate entry.
# tuple is a static.
# tuple is a collection of object type elements.
# tuple is a class of collection.
# Tuples are similar to lists, but unlike lists, they cannot be changed after their creation (i.e., they are immutable).
# Tuples can hold elements of different data types.
# The main characteristics of tuples are being ordered, heterogeneous and immutable.
# first element store by defalut in zero index.
```

```python
s=('ravi','kumar',66,33,12,'kapli','ravi')
print(s[0])
```

```python
s=('ravi','kumar',66,33,12,'kapli','ravi')
s[3]=456
print(s[3])
```

```python
# slicing
s=('ravi','kumar',66,33,12,'kapli','raj')
print(s[2:6])
print(s[::-1])
```

```python
s=('ravi','kumar',66,33,12,'kapli','ravi')
print(len(s))
```

```python
s=('ravi','kumar',66,'ravi',33,12,'kapli','raj','ravi')
for i in range(0,len(s)):
    if (s[i]=='ravi'):
        print(i)
```

```python
# Nested tuple
s=(4,3,2,'ravi','raj',88,89,99)
s2=(33,44,55,'ram','ramu')
s3=(s,s2)
print(s3)
```

```python
# Repat the tuple vlaues
t=(34,45,56,'ram','raj','ramu','sonu')
t2=(t*2)
print(t2)
```

```python
# Tuple unpacking
t=('ravi','arjun','arun',88,898,99,1009)
a,b,*c=t
print(a)
print(b)
print(c)
```

```python
# Tuple packing
t='ravi','arjun','arun',88,898,99,1009
print(t)
print(type(t))
```

```python
a,b,*c,d=(10,'hello',3.14,77,'raj',99)
print(a)
print(b)
print(*c)
print(d)
```

```python
# mathemathical operations it apply only when all the item of tuple is in numeric form.
t=(22,33,44,55,66,77,88,99,110)
print(sum(t))
print(max(t))
print(min(t))
print(len(t))
```

```python
# Enumerate tells us the index value
s=('ravi','kumar',66,33,12,'kapli','raj')
for i,j in enumerate(s):
    print(i," ",j)
```

```python
t=(1,2,3,4,5,22,33,44,55,66,77,88,99,110)
res= all(i>5 for i in t)      # if all are greater then 5 then true otherwise false.
print(res)
```

```python
t=(1,2,3,4,5,22,33,44,55,66,77,88,99,110)
res= any(i>5 for i in t)       # if anyone is greater than 5 then true otherwise false.
print(res)
```
