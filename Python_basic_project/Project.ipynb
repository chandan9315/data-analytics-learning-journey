```python
# simple small program for calculator using while loop.
while True:
    print("1.ADD")
    print("2.SUb")
    print("3.MUL")
    print("4.DIV")
    print("5.EXIT")

    x=int(input("enter the number"))

    if(x==1):
        a=int(input("enter the first value"))
        b=int(input("enter the second value"))
        c=a+b
        print(c)
    elif(x==2):
        a=int(input("enter the first value"))
        b=int(input("enter the second value"))
        c=a-b
        print(c)
    elif(x==3):
        a=int(input("enter the first value"))
        b=int(input("enter the second value"))
        c=a*b
        print(c)
    elif(x==4):
        a=int(input("enter the first value"))
        b=int(input("enter the second value"))
        c=a/b
        print(c)
    elif(x==5):
        break
    else:
        print("no choice")
        
```
