```python
# break : break is a keyword to terminate a loop according your condition.
# Continue
# pass:
```

```python
for i in range(0,20):
    if (i==5):
        print("This is my first break point")
        break
        print(i)
```

```python
for i in range(0,20):
    if (i==5):
        print("This is my first break point")
        continue
    elif(i==10):
        print("This is my first break point")
        break
        print(i)
```

```python
import time
for i in range(0,20):
    if (i==5):
        time.sleep(5)
        print("This is my first break point",i)
        continue
    elif(i==10):
        time.sleep(5)
        print("This is my second break point",i)
        continue
    elif(i==15):
        print("This is my third break point",i)
        break
    print(i)
```
