```python
# For print A,B,C,D,E we can use the ASCI code like A=65 then convert these code into character.
for i in range(65,71):
    for j in range(65,i+1):
        print(chr(j),end="")
    print()
```

```python
# For find the ASCI value of any variable. 
k='F'
p=ord(k)
print(p)
```
