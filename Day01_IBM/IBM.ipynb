```python
# Copy: In copy we create a new object instead of shearing memory.
a=[1,2,3]
b=a
b.append(5)
print(b)
print(a)
```

```python
original=[[1,2],3]
copy_list=original.copy()
copy_list[1]="CHANGED"
copy_list[0][1]=55
copy_list.append("NEW")              # shelo copy
print(original)
print(copy_list)
print(f"Original:{original}")
print(f"Copy list:{copy_list}")
```

```python
import copy
a=[[1,2],[2,4]]
b=copy.deepcopy(a)                  # Deep copy
b[0][0]=99
print(a)
print(b)
```

```python
# copy has two types:
# 1. shelo :
# 2. Deep :
```
