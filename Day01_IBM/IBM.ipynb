```python
#---------------------
# Data-Cleaning
#---------------------
```

```python
df=df.drop_duplicates()
```

```python
df.iloc[0,1]='Chandan'          # for change vlaues in columns.
df
```

```python
# Ensure the numeric columns
numeric_cols=['month_number', 'facecream', 'facewash', 'toothpaste', 'bathingsoap','shampoo', 'moisturizer', 'total_units', 'total_profit']
for col in numeric_cols:
    df[col]=pd.to_numeric(df[col],errors='coerce')
df
```

```python
# Filling any numerics vlaue with median.
for col in numeric_cols:
    df[col].fillna(df[col].median(),inplace=True)
df
```

```python
# for add a average column.
df['Average']=df['total_profit']/df['total_units']
df
```

```python
c=df.loc[df['total_units'].idxmax()]       # for print the filter row.
c
```

```python
top_profit=df[['month_number','total_profit']].sort_values(by='total_profit',ascending=False)
top_profit.head()              # for find the maximum selling unit.
```

```python
top_units=df[['month_number','total_units']].sort_values(by='total_units',ascending=False)
top_units.head() 
```

```python
d=df.loc[df['total_profit'].idxmax()]       # for print the filter row.
d
```

```python
Average=df['Average'].mean()
Average
```

```python
#sales tread over the months
monthly_unit = df.groupby('month_number')['total_units'].sum()
monthly_unit
monthly_profit = df.groupby('month_number')['total_profit'].sum()
monthly_profit
```

```python
#----------------------
# Matplotlib
#---------------------- 
```

```python
plt.figure(figsize=(10,5))
plt.plot(monthly_unit.index,monthly_unit.values,marker='*',color='green')
plt.title("Total Units sold per Month")
plt.xlabel('Month Number')
plt.ylabel('Unit Sold')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(monthly_profit.index,monthly_profit.values,marker='o')
plt.title("Total Units sold per Month")
plt.xlabel('Month Number')
plt.ylabel('Unit Sold')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(monthly_profit.index,monthly_profit.values)
plt.title("Total Units sold per Month")
plt.xlabel('Month Number')
plt.ylabel('Unit Sold')
plt.show()
```

```python
plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
plt.plot(monthly_profit.index,monthly_profit.values,marker='o')
plt.title("Total Units sold per Month")
plt.xlabel('Month Number')
plt.ylabel('Unit Sold')
plt.grid(True)

plt.subplot(1,2,2)
plt.bar(monthly_profit.index,monthly_profit.values)
plt.title("Total Units sold per Month")
plt.xlabel('Month Number')
plt.ylabel('Unit Sold')

plt.tight_layout()
plt.show()
```
